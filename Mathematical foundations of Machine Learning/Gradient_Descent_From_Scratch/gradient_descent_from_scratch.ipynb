# <a href='#intro'>1. Introduction</a>

In Machine Learning, models learn by minimizing a loss function (la fonction de perte) called Mean Squared Error (MSE).

# <a href='#import'>3. Generate Synthetic Data</a>


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

X = np.random.rand(100)
y = 4 * X + 3 + np.random.randn(100) * 0.2

plt.scatter(X, y)
plt.title("Synthetic Dataset")
plt.show()
```

# <a href='#cnn'>4. Implement Gradient Descent</a>


```python
w = 0
b = 0

lr = 0.1
epochs = 100

loss_history = []

n = len(X)

for epoch in range(epochs):

    y_pred = w * X + b

    loss = np.mean((y - y_pred) ** 2)
    loss_history.append(loss)

    dw = (-2 / n) * np.sum(X * (y - y_pred))
    db = (-2 / n) * np.sum(y - y_pred)

    w = w - lr * dw
    b = b - lr * db
```

---
The for loop is where the "learning" happens. It repeats these four steps for each epoch:
within this code w and b are simple THETA1 and THETA0
- **A. Prediction**
The model calculates its current guess for $y$ based on the current values of $w$ and $b$.

- **B. Loss Calculation**
This calculates the Mean Squared Error (MSE). It measures how far off the predictions are from the actual values. We square the differences so that large errors are penalized more heavily and negative values don't cancel out positive ones.

- **C. Gradient Calculation (Calculus)**
To improve the model, we need to know the direction and magnitude of the change needed for $w$ and $b$. These are the partial derivatives of the MSE loss function:

$dw$ tells us how much the loss changes if we change $w$.$db$ tells us how much the loss changes if we change $b$.

- **D. Updating Parameters**
We update $w$ and $b$ by subtracting a fraction (lr) of the gradient. We subtract because we want to move downhill toward the minimum error.

---


# <a href='#cnn'>5. Visualize Loss Convergence</a>


```python
plt.plot(loss_history)
plt.title("Loss Convergence")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

The model successfully learned the relationship between X and y.

The loss decreased progressively during training, demonstrating how gradient descent iteratively improves the model parameters.


# <a href='#concl'>6. Visualize the Final Model</a>


```python
y_pred = w * X + b

plt.scatter(X, y)
plt.plot(X, y_pred, color="red")
plt.title("Linear Regression Fit")
plt.show()
```
